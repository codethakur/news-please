```python
import sys
import time
from bs4 import BeautifulSoup
import requests
from newspaper import Article
import newspaper
import nltk
nltk.download('punkt')

url = 'https://www.worldbank.org/en/topic/indigenouspeoples#1'

r = requests.get(url)
htmlContent = r.content
soup = BeautifulSoup(htmlContent, 'html.parser')

article = Article(url, language="en")
article.download()
article.parse()
article.nlp() 
print('headline:', article.title)

print('lead paragraph:')
print(soup.find("p",article.__class__))

print("\n")
print("main Text:")
print(article.text)
print("Authors: Washington, D.C", article.authors)
print('main image', article.top_image)

#print(article.__class__)

print("\n")
print('Summry:\n',article.summary)

newspaper.languages()


```
